```python
import pandas as pd
import os
```

```python
# merge all correlation df, image df, and prompt df
```

```python
prompt_df_paths = [
    "./prompt_0.csv",
    "./prompt_5000.csv",
    "./prompt_10000.csv",
    "./prompt_15000.csv",
    "./prompt_20000.csv",
    "./prompt_25000.csv",
    "./prompt_30000.csv",
    "./prompt_35000.csv",
]

merged_df = pd.DataFrame()
for path in prompt_df_paths:
    df = pd.read_csv(path)
    merged_df = pd.concat([merged_df, df])

# merge row with same text and create a dictionary to map old id to new id
id_dict = {}
text_dict = {}

for i, row in merged_df.iterrows():
    id = row.id
    text = row.text
    if text in text_dict:
        id_dict[id] = text_dict[text]
    else:
        text_dict[text] = id
        id_dict[id] = id

# only keep rows with values in id_dict
merged_df = merged_df[merged_df.id.isin(text_dict.values())]
merged_df

# merged_df.to_csv("./prompt.csv", index=False)
```

```python
pairs_df_paths = [
    "./pairs_0.csv",
    "./pairs_5000.csv",
    "./pairs_10000.csv",
    "./pairs_15000.csv",
    "./pairs_20000.csv",
    "./pairs_25000.csv",
    "./pairs_30000.csv",
    "./pairs_35000.csv",
]

merged_df = pd.DataFrame()
for path in pairs_df_paths:
    df = pd.read_csv(path)
    merged_df = pd.concat([merged_df, df])

# map prompt_id to new id in id_dict
merged_df["prompt_id"] = merged_df["prompt_id"].map(id_dict)

merged_df.to_csv("./pairs.csv", index=False)
```

```python
correlation_df_paths = [
    "./correlation_0.csv",
    "./correlation_5000.csv",
    "./correlation_10000.csv",
    "./correlation_15000.csv",
    "./correlation_20000.csv",
    "./correlation_25000.csv",
    "./correlation_30000.csv",
    "./correlation_35000.csv",
]

merged_df = pd.DataFrame()
for path in correlation_df_paths:
    df = pd.read_csv(path)
    merged_df = pd.concat([merged_df, df])

# map prompt_id to new id in id_dict
merged_df["prompt_id"] = merged_df["prompt_id"].map(id_dict)

# merge rows with same prompt_id then concat image_id by space
merged_df = merged_df.groupby("prompt_id").agg(
    {"image_id": lambda x: " ".join(x.astype(str))}
)

# reset index
merged_df = merged_df.reset_index()

merged_df.to_csv("./correlation.csv", index=False)
```
